```python
#Faça um Programa que peça as quatro notas de 10 alunos, 
#calcule e armazene num vetor a média de cada aluno,
#imprima o número de alunos com média maior ou igual a 7.0.

notas = []
nota = 0
for i in range(0,10):
    for n in range(0,4):
        print(f"Aluno {i+1}")
        nota += float(input(f"Digite a {n+1}ª nota:"))
    nota = nota/4
    notas.append(nota)
    nota = 0

alunosAprovados = [n for n in notas if n>=7]
print("Quantidade de alunos Aprovados:", len(alunosAprovados))
print(alunosAprovados)
```

```python
#TEMOS 3 TIPOS DE OPERADORES

#<OPERADORES MATEMÁTICOS>
#soma(+), subtração(-), multiplicação(*), divisão(/), exponenciação(2**5), radiciação(2**(1/5), 
#divisão intera (//), resto de uma divisão inteira(%)

print("\nEXENPLOS DE OPERAÇÔES COM INT's\n")
print("Soma (+): ", 1 + 1)  # soma
print("Subtração (-): ",2 - 2)  # subtração
print("Multiplicação(*): ",2 * 2)  # multiplicação
print("Divisão (/): ", 2 / 2)  # divisão
print("Potenciação (**): ",3 ** 5)  # potenciação
print("Radiciação (**(n/m)):  ",4 ** (1/2))  # radiciação
print("Divisão Inteira (//): ",11//2) #retorna o resultado de uma divisão inteira
print("Resto de uma Divisão Inteira (%): ", 11 % 2)  # retorna o resto de uma divisão inteira
print("-------------------------------------------\n")

#OPERADORES RELACIONAIS
# Maior(>), Menor(<), Maior igual (>=), Menor Igual (<=), diferente(!=), igualdade (==)
# TEM COMO RESULTADO UM VERDADEIRO OU FALSO

print("1: ",1>2) #FALSE
print("2: ",2>1) #TRUE
print("3: ",2<0) #FALSE
print("4: ",1<3) #TRUE
print("5: ",5>=4) #TRUE
print("6: ",4>=4) #TRUE
print("7: ",5<=5) #TRUE
print("8: ",5<=6) #TRUE
print("9: ",5!=5) #FALSE
print("10: ",5!=4) #TRUE
print("11: ",5==5) #TRUE
print("12: ",5=="5") #FALSE
print("13: ","5"=="5") #TRUE
print("14: ",5==5.0) #TRUE
print("---------------------------------------\n")

#OPERADORES LÓGICOS
#E (AND), OU (OR), Negação(Not)
#RELAÇÃO DE DUAS OU MAIS RELAÇÕES

print("1: ", (4>2) and ("5"=="5")) #TRUE
print("2: ",(4<2) and ("5"=="5") ) #FALSE
print("3: ",(17>=10) and ("Oi"=="Oi")) #TRUE
print("4: ",(17>=10) and ("Oi"=="oi")) #FALSE
print("5: ",(4<2) or ("5"=="5")) #TRUE
print("6: ",(7!=7) or ("Jaime"=="jaime")) #FALSE
print("7: ",2<10 or 10<50) #TRUE
print("8: ",not(7!=7)) #TRUE
print("9: ",not(8!=7)) #FALSE
print("--------------------------------------\n")

#OUTROS EXEMPLOS
print("1: ", 10/5==5) #FALSE
print("2: ", 10/2==5) #TRUE
print("3: ", 11%2==1) #TRUE
print("4: ", 25**(1/2)>=5) #TRUE
print("5: ", not(25**(1/2)>=5)) #FALSE
print("6: ",not(10/5==5)) #TRUE
```

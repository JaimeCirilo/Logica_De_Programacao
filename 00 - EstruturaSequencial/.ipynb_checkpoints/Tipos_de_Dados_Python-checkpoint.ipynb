```python
#Aula python para turma 4 de Introdução a Linguagem de Programação Python!

# TIPOS PRIMITIVOS
####### STRING ou TEXTO#######
# - O QUE A GENTE PODE ESCREVER? TUDO QUE ESTIVER DENTRO DE ASPAS SIMPLES OU DUPLAS É CONSIDERADO TEXTO.

# EXEMPLOS DE STRINGS DENTRO DO COMANDO PRINT().
print("\nEXEMPLOS DE STRINGS\n")
print("Olá pessoal! Tudo bem com vocês? Essa é a aula de PYTHON")
print("a")
print("A")
print("9")
print("200.72")
print("")
print('')


# EXEMPLOS DE OPERAÇÕES COM STRINGS DENTRO DO COMANDO PRINT().
print("\nEXEMPLOS DE OPERAÇÕES COM STRINGS\n")
print("Oi" + " pessoal")  # concatenação de strings
print("2" + "2")  # concatenação de strings
print("1" * 3)  # repetição de strings

print("\n--------------------------------------\n")

####### INT OU INTEIRO#######
# - O QUE A GENTE PODE ESCREVER? QUALQUER NÚMERO QUE NÃO CONTENHA PONTO OU CASA DECIMAL.

# EXEMPLOS DE INT DENTRO DO COMANDO PRINT().
print("\nEXEMPLOS DE INT's\n")
print(1)
print(2)
print(3)
print(1000)


soma_numeros = 12313 + 2131321

# EXEMPLOS DE OPERAÇÕES COM INTs DENTRO DO COMANDO PRINT().
print("\nEXEMPLOS DE OPERAÇÔES COM INT's\n")
print("Soma (+): ", soma_numeros)  #soma
print("Subtração (-): ",2 - 2)  #subtração
print("Multiplicação(*): ",2 * 2)  #multiplicação
print("Divisão (/): ", 2 / 2)  # divisão
print("Potenciação (**): ",3 ** 5)  #potenciação
print("Radiciação (**(n/m)):  ",4 ** (1/2))  #radiciação
print("Divisão Inteira (//): ",11//2) #retorna o resultado de uma divisão inteira
print("Resto de uma Divisão Inteira (%): ", 11 % 2)  #retorna o resto de uma divisão inteira


print("\n--------------------------------------\n")

# EXEMPLOS DE OPERAÇÕES COM STRINGS DENTRO DO COMANDO PRINT().
print("\nEXEMPLOS DE OPERAÇÕES COM STRINGS\n")
print("Oi" + " pessoal")  # concatenação de strings
print("2" + "2")  # concatenação de strings
print("1" * 3)  # repetição de strings



####### FLOAT OU PONTO FLUTUANTE#######
# - O QUE A GENTE PODE ESCREVER? QUALQUER NÚMERO QUE CONTENHA PONTO OU CASA DECIMAL.
print("\nEXEMPLOS DE OPERAÇÔES COM FLOAT's\n")
print(1.0 + 1.0)  # soma
print(2.0 - 2.0)  # subtração
print(2.0 * 2.0)  # multiplicação
print(2.0 / 2.0)  # divisão
print(3.0 ** 5.0)  # potenciação
print(4.0 ** (1/2))  # radiciação
print(11.0 % 2.0)  # retorna o resto de uma divisão inteira

# BOOL OU BOOLEANO (BINÁRIO)
# - O QUE A GENTE PODE ESCREVER? True ou False ; 1 ou 0.
print("\nEXEMPLOS DE BOOL's\n")
print(True)
print(False)
print(bool(0))
print(bool(1))
```
